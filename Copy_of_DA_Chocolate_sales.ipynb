# Reading tables


```python

calendar_df = pd.read_csv('/content/drive/MyDrive/Chocolate/calendar.csv')
print("calender headers: \n", calendar_df.head(), '\n', calendar_df.shape)

customers_df = pd.read_csv('/content/drive/MyDrive/Chocolate/customers.csv')
print("customers headers: \n", customers_df.head(), '\n', customers_df.shape)

product_df = pd.read_csv('/content/drive/MyDrive/Chocolate/products.csv')
print("products headers: \n", product_df.head(), '\n', product_df.shape)

sales_df = pd.read_csv('/content/drive/MyDrive/Chocolate/sales.csv')
print("sales headers: \n", sales_df.head(), '\n', sales_df.shape)

stores_df = pd.read_csv('/content/drive/MyDrive/Chocolate/stores.csv')
print("stores headers: \n", stores_df.head(), "\n", stores_df.shape)
```

# Total profit

```python
total_profit = sales_df['profit'].sum()
print(f"Total profit is : {round(total_profit,2):,}")
```

```python
total_revenue = sales_df['revenue'].sum()
print(f"Total revenue is : {round(total_revenue,2):,}")
```

```python
avg_choco_per_order =  sales_df['quantity'].mean()
print(f"AVG chocolates per order is: {round(avg_choco_per_order)}")
```

```python
avg_profit_per_order =  sales_df['profit'].mean()
print(f"AVG profit per orderis: {round(avg_profit_per_order,2)}")
```

```python
Gross_profit_margin = (total_profit/ total_revenue) * 100
print("Gross Profit Margin:", round(Gross_profit_margin, 2), "%")
```

```python
merged_df = pd.merge(sales_df, stores_df, on='store_id', how='left')
```

```python
out = (merged_df.groupby('country', as_index=False)['quantity'].sum())

print("before sort", out)

result = out.sort_values(by='quantity', ascending=False)

print(f"out:\n {result[["country", "quantity"]]}")
print(f"top 2 rows:\n {result[:2]["country"]}")


print("out:\n", result[["country", "quantity"]].to_string(index=False))
print("\n \n top 2 rows:\n", result.head(2)["country"].to_string(index=False))
```

```python
out = (merged_df.groupby('country', as_index=False)['quantity'].sum())
result = out.sort_values(by='quantity', ascending=False)

print("out:\n", result[["country", "quantity"]].to_string(index=False))
print("\n \n top 2 rows:\n", result.head(2)["country"].to_string(index=False))
```
